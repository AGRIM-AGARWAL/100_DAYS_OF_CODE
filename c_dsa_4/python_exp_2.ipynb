```python
#Question 3 : leap year check
x = int(input("year : "))
if x%4==0 :
    if(x%100==0 and x%400!=0) :
        print("not leap year")
    elif x%400==0 :
        print("leap year")
    else :
        print("leap year")
else :
    print("not a leap year")
```
